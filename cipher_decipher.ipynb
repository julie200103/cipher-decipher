---
Caesar Shift / Fixed Shift

---

```python
def caesar_cipher(text, shift=3):
    return ''.join(chr((ord(c) - 97 + shift) % 26 + 97) if 'a' <= c <= 'z' else c for c in text)

def caesar_decipher(text, shift=3):
    return caesar_cipher(text, -shift)

print(caesar_cipher("hello"))  # khoor
print(caesar_decipher("khoor"))  # hello
```

```python
print(caesar_cipher("kshama"))
print(caesar_decipher("nvkdpd"))
```

---
Reverse Mapping

---

```python
def reverse_mapping(text):
    return ''.join(chr(122 - (ord(c) - 97)) if 'a' <= c <= 'z' else c for c in text)

print(reverse_mapping("hello"))  # svool
print(reverse_mapping("svool"))  # hello (Same function works both ways!)
```

---
Vowel Consonant Swap

---

```python
def vowel_consonant_swap(text):
    vowels = 'aeiou'
    consonants = 'bcdfghjklmnpqrstvwxyz'

    def swap(c):
        if c in vowels:
            return vowels[(vowels.index(c) + 1) % len(vowels)]
        elif c in consonants:
            return consonants[(consonants.index(c) + 1) % len(consonants)]
        return c

    return ''.join(swap(c) for c in text)

print(vowel_consonant_swap("coco"))  # jimmu
print(vowel_consonant_swap("bruno"))  # hello (same function reverses it)
```

---
Ascii Increment Decrement

---

```python
def ascii_swap(text):
    return ''.join(chr(ord(c) + 1) if i % 2 == 0 else chr(ord(c) - 1) for i, c in enumerate(text))

def ascii_unswap(text):
    return ascii_swap(text)  # Since it's just a +1/-1 swap, applying again reverses it.

print(ascii_swap("hello"))  # "idmln"
print(ascii_unswap("idmln"))  # "hello"
```

---
Mirror Ascii Transformation

---

```python
def mirror_ascii(text):
    return ''.join(chr(219 - ord(c)) if 'a' <= c <= 'z' else c for c in text)

print(mirror_ascii("hello"))  # svool
print(mirror_ascii("svool"))  # hello (same function for both!)
```
